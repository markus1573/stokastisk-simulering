```python
import numpy as np

n = 20

angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
cities = np.column_stack((np.cos(angles), np.sin(angles)))


current_route = np.arange(0,n,1)
np.random.shuffle(current_route)

cost_history = []


def calculate_cost(route, cities):
    total_dist = 0
    for i in range(0, n-1):
        city_a = cities[route[i]]
        city_b = cities[route[i+1]]

        total_dist += np.linalg.norm(city_a - city_b)
    total_dist += np.linalg.norm(cities[route[0]] - cities[route[-1]])

    return total_dist



import matplotlib.pyplot as plt
itert = 10000
current_cost = calculate_cost(current_route, cities)


for i in range(itert):
    proposed_route = current_route.copy()
    idx1, idx2 = np.random.choice(n, size = 2, replace = False)
    proposed_route[idx1], proposed_route[idx2] = proposed_route[idx2], proposed_route[idx1]

    T_k = 1/np.sqrt(1 + i)

    proposed_cost = calculate_cost(proposed_route, cities)

    delta_f = proposed_cost - current_cost

    if delta_f <= 0:
        current_route = proposed_route
        current_cost = proposed_cost

    else:
        alpha = np.exp(-(delta_f)/T_k)
        if np.random.uniform(0,1) < alpha:
            current_route = proposed_route
            current_cost = proposed_cost

    cost_history.append(current_cost)



optimized_path = cities[current_route]
optimized_path = np.vstack((optimized_path, optimized_path[0]))

plt.figure(figsize=(6, 6))
plt.plot(optimized_path[:, 0], optimized_path[:, 1], marker='o', linestyle='-', color='b')
plt.title(f"Simulated Annealing TSP (Sanity Check)\nFinal Cost: {current_cost:.2f}")
plt.grid(True)
plt.axis('equal')
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(cost_history, color='r', linewidth=1.5)
plt.title("Simulated Annealing Convergence")
plt.xlabel("Iteration")
plt.ylabel("Path Cost")
plt.grid(True)
plt.show()

```

```python
import numpy as np


def bootstrap_median_analysis(sample, k=100):

    n = len(sample)

    sample_mean = np.mean(sample)
    sample_median = np.median(sample)

    bootstrap_samples = np.random.choice(sample, size=(k, n), replace=True)

    bootstrap_means = np.mean(bootstrap_samples, axis=1)
    bootstrap_medians = np.median(bootstrap_samples, axis=1)

    var_bootstrap_mean = np.var(bootstrap_means, ddof=1)
    var_bootstrap_median = np.var(bootstrap_medians, ddof=1)

    return {
        "original_mean": sample_mean,
        "original_median": sample_median,
        "var_bootstrap_mean": var_bootstrap_mean,
        "var_bootstrap_median": var_bootstrap_median,
    }


# --- Simulation Setup ---
np.random.seed(42)
n = 200
beta = 1
k_pareto = 1.05
k = 100

U = np.random.uniform(0, 1, n)
Xi = beta * (U ** (-1 / k_pareto))

results = bootstrap_median_analysis(Xi, k=k)

print(f"1. Original Sample Mean:   {results['original_mean']:.4f}")
print(f"   Original Sample Median: {results['original_median']:.4f}")
print(
    f"2. Bootstrap Variance of the Mean:   {results['var_bootstrap_mean']:.4f}"
)
print(
    f"3. Bootstrap Variance of the Median: {results['var_bootstrap_median']:.4f}"
)
```
